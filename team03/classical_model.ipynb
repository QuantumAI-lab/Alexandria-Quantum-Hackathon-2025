```python
# Define the reaction coordinate: distance between N and C atoms
def create_molecule(n_c_distance):
    # NH₃ molecule
    nh3 = Atoms('NH3',
                positions=[(0.0000, 0.0000, 0.0000),
                          (0.0000, 0.9395, 0.0000),
                          (0.8137, -0.4697, 0.0000),
                          (-0.8137, -0.4697, 0.0000)])

    # CO₂ molecule (aligned along x-axis)
    co2 = Atoms('CO2',
                positions=[(n_c_distance, 0.0000, 0.0000),
                          (n_c_distance + 1.1600, 0.0000, 0.0000),
                          (n_c_distance - 1.1600, 0.0000, 0.0000)])

    # Combine molecules
    system = nh3 + co2
    return system

# Set up DFT calculation
def calculate_energy(atoms):
    # Convert ASE atoms to PySCF format
    positions = atoms.get_positions()
    symbols = atoms.get_chemical_symbols()
    atom_str = []
    for i, sym in enumerate(symbols):
        atom_str.append(f"{sym} {positions[i][0]} {positions[i][1]} {positions[i][2]}")

    # Create molecule object
    mol = gto.M(atom=atom_str, basis = 'def2-TZVP')  # Using smaller basis for speed

    # Perform DFT calculation
    mf = dft.RKS(mol)
    mf.xc = 'b3lyp'
    energy = mf.kernel()

    return energy

# Perform PES scan
n_c_distances = np.linspace(1.8, 3.2, 8)  # Reduced points for faster computation
energies = []

print("Starting PES scan...")
for i, dist in enumerate(n_c_distances):
    print(f"Step {i+1}/{len(n_c_distances)}: N-C distance = {dist:.2f} Å")

    # Create initial geometry
    system = create_molecule(dist)

    # Calculate energy directly without optimization
    try:
        energy = calculate_energy(system)
        energies.append(energy)
        print(f"  Energy = {energy:.6f} Hartree")
    except Exception as e:
        print(f"  Calculation failed: {e}")
        energies.append(np.nan)

print("PES scan completed!")
```

```python
import numpy as np
from ase import Atoms
from pyscf import gto, dft
import matplotlib.pyplot as plt

# ===== Helper function to calculate energy using PySCF =====
def calculate_energy(atoms, basis='def2-TZVP'):
    positions = atoms.get_positions()
    symbols = atoms.get_chemical_symbols()
    atom_str = []
    for i, sym in enumerate(symbols):
        atom_str.append(f"{sym} {positions[i][0]} {positions[i][1]} {positions[i][2]}")

    mol = gto.M(atom=atom_str, basis=basis)
    mf = dft.RKS(mol)
    mf.xc = 'b3lyp'
    energy = mf.kernel()
    return energy

# ===== Define molecules =====
# NH3 geometry
nh3 = Atoms('NH3',
            positions=[(0.0000, 0.0000, 0.0000),
                      (0.0000, 0.9395, 0.0000),
                      (0.8137, -0.4697, 0.0000),
                      (-0.8137, -0.4697, 0.0000)])

# CO2 geometry (linear along x-axis, centered at 0)
co2 = Atoms('CO2',
            positions=[(0.0000, 0.0000, 0.0000),
                      (1.1600, 0.0000, 0.0000),
                      (-1.1600, 0.0000, 0.0000)])

# ===== Compute isolated molecule energies =====
print("Computing isolated molecule energies...")
E_NH3 = calculate_energy(nh3)
E_CO2 = calculate_energy(co2)
print(f"Energy NH3 = {E_NH3:.6f} Hartree")
print(f"Energy CO2 = {E_CO2:.6f} Hartree")

# ===== Function to build NH3 + CO2 system with variable N-C distance =====
def create_complex(n_c_distance):
    # NH3 at origin (same as above)
    nh3 = Atoms('NH3',
                positions=[(0.0000, 0.0000, 0.0000),
                          (0.0000, 0.9395, 0.0000),
                          (0.8137, -0.4697, 0.0000),
                          (-0.8137, -0.4697, 0.0000)])
    # CO2 shifted along x-axis
    co2 = Atoms('CO2',
                positions=[(n_c_distance, 0.0000, 0.0000),
                          (n_c_distance + 1.1600, 0.0000, 0.0000),
                          (n_c_distance - 1.1600, 0.0000, 0.0000)])
    return nh3 + co2

# ===== PES scan for binding energy =====
n_c_distances = np.linspace(1.8, 3.2, 8)
binding_energies = []

print("\nStarting PES scan for binding energies...")
for dist in n_c_distances:
    system = create_complex(dist)
    E_complex = calculate_energy(system)
    # Binding energy = E_complex - (E_NH3 + E_CO2)
    E_bind = E_complex - (E_NH3 + E_CO2)
    binding_energies.append(E_bind)
    print(f"N-C distance = {dist:.2f} Å, Binding Energy = {E_bind:.6f} Hartree")
```

```python
import numpy as np
import matplotlib.pyplot as plt

complex_energies = np.array([
    -240.844828971281,
    -242.655750155489,
    -243.537346495117,
    -244.159866689913,
    -244.637576233054,
    -244.927046045637,
    -245.083617332531,
    -245.164433307442
])

E_NH3 = -56.5683081931929
E_CO2 = -188.669935659863

n_c_distances = np.linspace(1.8, 3.2, len(complex_energies))

binding_hartree = complex_energies - (E_NH3 + E_CO2)

hartree_to_kjmol = 2625.4996
hartree_to_kcalmol = 627.509474
binding_kjmol = binding_hartree * hartree_to_kjmol
binding_kcalmol = binding_hartree * hartree_to_kcalmol

print("Distance(Å)   E_complex(Hartree)    Binding(Hartree)    Binding(kJ/mol)   Binding(kcal/mol)")
for d, E_c, b_h, b_kj, b_kcal in zip(n_c_distances, complex_energies, binding_hartree, binding_kjmol, binding_kcalmol):
    print(f"{d:6.2f}        {E_c:14.6f}        {b_h:12.6f}        {b_kj:12.2f}        {b_kcal:12.2f}")

min_idx = np.nanargmin(binding_kjmol)
min_dist = n_c_distances[min_idx]
min_bind_kj = binding_kjmol[min_idx]
min_bind_kcal = binding_kcalmol[min_idx]

print("\nBest binding (minimum):")
print(f"  Distance = {min_dist:.2f} Å")
print(f"  Binding = {min_bind_kj:.2f} kJ/mol   ({min_bind_kcal:.2f} kcal/mol)")

# رسم
plt.figure(figsize=(7,4))
plt.plot(n_c_distances, binding_kjmol, 'o-', label='Binding energy (kJ/mol)')
plt.axhline(0, color='gray', linestyle='--')
plt.scatter([min_dist], [min_bind_kj], color='red', zorder=5, label=f"Min = {min_bind_kj:.2f} kJ/mol at {min_dist:.2f} Å")
plt.xlabel("N–C distance (Å)")
plt.ylabel("Binding energy (kJ/mol)")
plt.title("NH3–CO2 Binding Energy vs N–C distance")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyscf import gto, scf

def build_complex(R):
    mol = gto.Mole()
    mol.atom = f'''
    C 0.0 0.0 0.0
    O 0.0 0.0 1.16
    O 0.0 0.0 -1.16
    N 0.0 0.0 {R}
    H 0.0 0.94 {R+0.3}
    H 0.81 -0.47 {R+0.3}
    H -0.81 -0.47 {R+0.3}
    '''
    mol.basis = "sto-3g"
    mol.build()
    return mol

energies = []
distances = np.linspace(2.0, 5.0, 10)
for R in distances:
    mol = build_complex(R)
    mf = scf.RHF(mol).run()
    energies.append(mf.e_tot)

hartree_to_kcal = 627.509
energies_kcal = [(E - min(energies)) * hartree_to_kcal for E in energies]

plt.plot(distances, energies, marker='o', label="Energy (Hartree)")
plt.xlabel("Distance (Å)")
plt.ylabel("Energy (Hartree)")
plt.legend()
plt.show()

plt.plot(distances, energies_kcal, marker='s', color="red", label="Relative Energy (kcal/mol)")
plt.xlabel("Distance (Å)")
plt.ylabel("Energy (kcal/mol)")
plt.legend()
plt.show()

import pandas as pd
df = pd.DataFrame({"Distance (Å)": distances, "Energy (Hartree)": energies, "Relative Energy (kcal/mol)": energies_kcal})
print(df)
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from pyscf import gto, scf, dft

# ثوابت فيزيائية
kB = 1.380649e-23           # J/K
h = 6.62607015e-34          # J s
Na = 6.02214076e23          # 1/mol
c = 299792458.0             # m/s
R = 8.31446261815324        # J/(mol K)
hartree_to_joule = 4.3597447222071e-18   # J
hartree_to_kjmol = hartree_to_joule * Na / 1000.0
hartree_to_kcalmol = hartree_to_joule * Na / 4184.0
amu_to_kg = 1.66053906660e-27

# كتل الذرات
atomic_masses_amu = {
    'H': 1.00782503223,
    'C': 12.0,
    'N': 14.00307400443,
    'O': 15.99491461957
}

# ===============================
# دوال مساعدة
# ===============================

def guess_symmetry_number(mol):
    symbols = [mol.atom_symbol(i) for i in range(mol.natm)]
    if symbols.count('N')==1 and symbols.count('H')==3:
        return 3
    if symbols.count('C')==1 and symbols.count('O')==2:
        return 2
    return 1

def molecule_mass_kg(mol):
    masses = [atomic_masses_amu.get(mol.atom_symbol(i), 12.0) for i in range(mol.natm)]
    return sum(masses) * amu_to_kg

def principal_moments_kgm2(mol):
    coords = mol.atom_coords() * 1.0e-10  # Å -> m
    masses = np.array([atomic_masses_amu.get(mol.atom_symbol(i), 12.0) for i in range(mol.natm)]) * amu_to_kg
    cm = np.sum(coords * masses[:,None], axis=0) / np.sum(masses)
    r = coords - cm
    I = np.zeros((3,3))
    for i in range(mol.natm):
        m = masses[i]
        I += m * (np.dot(r[i], r[i]) * np.eye(3) - np.outer(r[i], r[i]))
    eigvals = np.linalg.eigvalsh(I)
    eigvals.sort()
    return eigvals

def is_linear(mol, tol_ratio=1e-3):
    I = principal_moments_kgm2(mol)
    if I[0] < tol_ratio * max(I[1], I[2]):
        return True
    return False

def extract_pos_freqs_from_analysis(analysis):
    freqs = np.array(analysis['freq_wavenumber'], dtype=complex)
    pos = []
    for f in freqs:
        if np.iscomplex(f):
            if abs(f.imag) > 1e-6:
                continue
            f = f.real
        if f > 1.0:
            pos.append(float(f))
    return np.array(pos)

def thermo_from_analysis(mf, mol, analysis, T=298.15, P=101325.0, sigma_override=None):
    """
    حساب الخصائص الحرارية للجزيء: طاقة حرارية، انتروبي، Gibbs
    """
    # استخراج الترددات
    freqs = extract_pos_freqs_from_analysis(analysis)
    # عدد الاهتزازات
    Nvib = len(freqs)

    # ثابت الغاز
    kbJ = kB  # J/K

    # طاقة اهتزازية لكل تردد
    energies = h * c * 100 * freqs  # تحويل cm^-1 -> J

    # طاقة اهتزازية متوسطة لكل اهتزاز
    ZPE = 0.5 * np.sum(energies)  # J

    # الانتروبي والحرارة النوعية (تقريب هارمونك)
    vibrational_contrib = energies / (kbJ * T)
    S_vib = R * np.sum(vibrational_contrib / (np.exp(vibrational_contrib) - 1) - np.log(1 - np.exp(-vibrational_contrib)))

    # الطاقة الداخلية
    U_vib = np.sum(energies / (np.exp(vibrational_contrib) - 1) + 0.5 * energies)  # J

    # طاقة Gibbs الحرارية
    G = U_vib - T * S_vib  # J

    return {'ZPE': ZPE, 'S_vib': S_vib, 'U_vib': U_vib, 'G': G}

# ===============================
# مثال: NH3 + CO2
# ===============================

def build_complex(R):
    mol = gto.Mole()
    mol.atom = f'''
    C 0.0 0.0 0.0
    O 0.0 0.0 1.16
    O 0.0 0.0 -1.16
    N 0.0 0.0 {R}
    H 0.0 0.94 {R+0.3}
    H 0.81 -0.47 {R+0.3}
    H -0.81 -0.47 {R+0.3}
    '''
    mol.basis = "sto-3g"
    mol.build()
    return mol

# ===============================
# مسح الطاقة
# ===============================
distances = np.linspace(2.0, 5.0, 10)
energies = []

for R in distances:
    mol = build_complex(R)
    mf = scf.RHF(mol).run()
    energies.append(mf.e_tot)

hartree_to_kcal = 627.509
energies_kcal = [(E - min(energies)) * hartree_to_kcal for E in energies]

# رسم منحنى الطاقة
plt.figure(figsize=(7,4))
plt.plot(distances, energies_kcal, 'o-', color='red', label="Relative Energy (kcal/mol)")
plt.xlabel("Distance N-C (Å)")
plt.ylabel("Energy (kcal/mol)")
plt.title("NH3–CO2 Energy vs N-C distance")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# طباعة النتائج
import pandas as pd
df = pd.DataFrame({"Distance (Å)": distances, "Energy (Hartree)": energies, "Relative Energy (kcal/mol)": energies_kcal})
print(df)
```
